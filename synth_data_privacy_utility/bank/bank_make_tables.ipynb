```python
import pandas as pd
import json
import os


output_dir = "data/data_outputs_bank"  # adjust if needed

utility_path = f"{output_dir}/utility_results_bank.csv"

mia_path = f"{output_dir}/model_mia_auc_bank.json"
worst_case_path = f"{output_dir}/worst_case_mia_auc_bank.json"


utility_df = pd.read_csv(utility_path)

with open(mia_path, "r") as f:
    model_mia_aucs = json.load(f)
with open(worst_case_path, "r") as f:
    worst_case_aucs = json.load(f)


# Summary table
rf_df = utility_df[utility_df["Model"] == "RandomForest"].copy()
rf_df["Privacy (MIA AUC)"] = rf_df["Trained_On"].str.lower().map(model_mia_aucs)
rf_df["Worst-Case MIA AUC"] = rf_df["Trained_On"].str.lower().map(worst_case_aucs)



summary_df = rf_df[["Trained_On", "Accuracy", "F1", "AUC", "Privacy (MIA AUC)", "Worst-Case MIA AUC"]]
summary_df = summary_df.rename(columns={
    "Trained_On": "Model",
    "AUC": "ROC AUC",
    "F1": "F1 Score"
})

summary_df["Model"] = summary_df["Model"].str.upper()
summary_df = summary_df.sort_values(by="Accuracy", ascending=False).reset_index(drop=True)




# Print tables
print("\n--- Bank Dataset: Privacy–Utility Summary (Random Forest) ---")
print(summary_df.to_string(index=False))



# summary_df.to_csv(f"{output_dir}/summary_table_bank.csv", index=False)
# print(summary_df.to_latex(index=False))
```
